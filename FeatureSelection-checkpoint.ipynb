```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

## Import data

```python
file_path = 'https://raw.githubusercontent.com/sguttula/KobeBryant-ShotSelection/main/data.csv'

df = pd.read_csv(file_path)
print(f'Features counts: {len(df.columns)}')
print(df.columns)
df.head()
```

## features lat / lon and loc_x / loc_y basically showing the same data indication (shape of basketball court)

```python
plt.figure(figsize=(20,8))

plt.subplot(121)
plt.scatter(df.lon, df.lat, alpha=0.02, color='blue')
plt.title('lat and lon')
plt.subplot(122)
plt.title('Location x and y')
plt.scatter(df.loc_x, df.loc_y, alpha=0.02, color='green')
plt.show()
```

## Create new columns

### 1. distance and angle

```python
# to calculate the polar cordinates
df['dist'] = np.sqrt((df.loc_x)**2 + (df.loc_y)**2)

loc_x_zero = df.loc_x == 0
df['angle'] = np.array([0] * len(df))
df['angle'][~loc_x_zero] = np.arctan(df['loc_y'][~loc_x_zero] / df['loc_x'][~loc_x_zero])
df['angle'][loc_x_zero] = np.pi/2\
# with the new distance and angle columns, we can drop lat and long columns
df.iloc[:,::-1].head()
```

```python
plt.figure(figsize=(18,5))
sns.regplot(data=df, x='dist', y='shot_distance', color='green', marker='x')
plt.show()
# same info contain
```

### 2. remanining minute and second

```python
df['time_remaining'] = df.minutes_remaining * 60 + df.seconds_remaining
df.iloc[:,::-1].head()
```

### 5. Shot zone area, shot zone basic and shot zone range

```python
df[['shot_zone_area', 'shot_zone_basic', 'shot_zone_range']].head()
```

```python
print(f'shot_zone_area:\n {df.shot_zone_area.unique()}\n')
print(f'shot_zone_basic:\n {df.shot_zone_basic.unique()}\n')
print(f'shot_zone_range:\n {df.shot_zone_range.unique()}')
```

```python
def scatter_plot_by_category(feat):
    gs = df.groupby(feat)
    cs = cm.rainbow(np.linspace(0, 1, len(gs)))
    for g, c in zip(gs, cs):
        plt.scatter(g[1].loc_x, g[1].loc_y, color=c, alpha=0.1)

plt.figure(figsize=(15,8))

plt.subplot(131)
scatter_plot_by_category(df.shot_zone_area)  
plt.title('shot_zone_area')

plt.subplot(132)
scatter_plot_by_category(df.shot_zone_basic)
plt.title('shot_zone_basic')

plt.subplot(133)
scatter_plot_by_category(df.shot_zone_range)
plt.title('shot_zone_range')
```
